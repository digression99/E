```python
# imputation missing values
impute_value = train['Age'].mean() # or median() 
train['Age'] = train['Age'].fillna(impute_value)
```

```python
# creating new variable. 

## for sex
# trereating as indicator -> {0,1} , 1:female
train['female'] = (train['Sex'] == 'female').astype(int)
train['female'][:10]
```

```python
train["child"] = train['Age'] < 14
```

```python
# building predictors 
predictors = ['Pclass','female', 'Age', 'child','Fare']
```

```python
train,test = train.random_split(0.8, seed=0)
```

```python
model = graphlab.logistic_classifier.create(train,
                                            target="Survived",
                                            features = predictors,
                                            validation_set = None)
```

```python
model.evaluate(test,metric='roc_curve')
```
